```python
"""
spearman correlation vs.  linear model coefficient
"""
import numpy as np
import pandas as pd

model_coeff = np.loadtxt("o38_coeff.txt")
spearman_correlation = pd.read_csv("o38_spearman.csv").fillna(value=0)  # make nan correlation to 0

spearman_correlation_matrix = np.zeros(model_coeff.shape)
spearman_correlation_matrix[0,:] = spearman_correlation["age"].drop(0).values[:-1]
spearman_correlation_matrix[1,:] = spearman_correlation["TSH"].drop(1).values[:-1]
spearman_correlation_matrix[2,:] = spearman_correlation["T3"].drop(2).values[:-1]
spearman_correlation_matrix[3,:] = spearman_correlation["TT4"].drop(3).values[:-1]
spearman_correlation_matrix[4,:] = spearman_correlation["T4U"].drop(4).values[:-1]
spearman_correlation_matrix[5,:] = spearman_correlation["FTI"].drop(5).values[:-1]


```

```python
from scipy.stats import pearsonr

print pearsonr(spearman_correlation_matrix[0,:], model_coeff[0,:])
print pearsonr(spearman_correlation_matrix[1,:], model_coeff[1,:])
print pearsonr(spearman_correlation_matrix[2,:], model_coeff[2,:])
print pearsonr(spearman_correlation_matrix[3,:], model_coeff[3,:])
print pearsonr(spearman_correlation_matrix[4,:], model_coeff[4,:])
print pearsonr(spearman_correlation_matrix[5,:], model_coeff[5,:])
```

```python
%matplotlib inline

# bellowing are the plot that spearman correlation as X-axis; model coeff as Y-axis
import matplotlib.pyplot as plt
plt.plot(spearman_correlation_matrix[0,:], model_coeff[0,:], 'ro')
plt.show()
```
