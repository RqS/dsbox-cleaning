```python
"""
spearman correlation vs.  linear model coefficient
"""
import numpy as np
import pandas as pd
from sklearn.preprocessing import scale

model_coeff = np.loadtxt("o38_scaled_coeff.txt")
pearson_correlation = pd.read_csv("o38_scaled_pearson.csv").fillna(value=0)  # make nan correlation to 0
spearman_correlation = pd.read_csv("o38_scaled_spearman.csv").fillna(value=0)  # make nan correlation to 0

spearman_correlation_matrix = np.zeros(model_coeff.shape)
spearman_correlation_matrix[0,:] = spearman_correlation["age"].drop(0).values[:-1]
spearman_correlation_matrix[1,:] = spearman_correlation["TSH"].drop(1).values[:-1]
spearman_correlation_matrix[2,:] = spearman_correlation["T3"].drop(2).values[:-1]
spearman_correlation_matrix[3,:] = spearman_correlation["TT4"].drop(3).values[:-1]
spearman_correlation_matrix[4,:] = spearman_correlation["T4U"].drop(4).values[:-1]
spearman_correlation_matrix[5,:] = spearman_correlation["FTI"].drop(5).values[:-1]

pearson_correlation_matrix = np.zeros(model_coeff.shape)
pearson_correlation_matrix[0,:] = pearson_correlation["age"].drop(0).values[:-1]
pearson_correlation_matrix[1,:] = pearson_correlation["TSH"].drop(1).values[:-1]
pearson_correlation_matrix[2,:] = pearson_correlation["T3"].drop(2).values[:-1]
pearson_correlation_matrix[3,:] = pearson_correlation["TT4"].drop(3).values[:-1]
pearson_correlation_matrix[4,:] = pearson_correlation["T4U"].drop(4).values[:-1]
pearson_correlation_matrix[5,:] = pearson_correlation["FTI"].drop(5).values[:-1]
```

```python
norm_corff = np.copy(model_coeff)
norm_corff=scale(norm_corff,axis=1)

top5 = np.zeros([6,6])
np.fill_diagonal(top5,1)

for i in range(top5.shape[1]):
    top5[i,top5[i,:]==0] = norm_corff[i,0:5]



np.savetxt("top5.txt",top5,fmt='%.3f',)

```

```python
from scipy.stats import pearsonr

print pearsonr(spearman_correlation_matrix[0,:], model_coeff[0,:])
print pearsonr(spearman_correlation_matrix[1,:], model_coeff[1,:])
print pearsonr(spearman_correlation_matrix[2,:], model_coeff[2,:])
print pearsonr(spearman_correlation_matrix[3,:], model_coeff[3,:])
print pearsonr(spearman_correlation_matrix[4,:], model_coeff[4,:])
print pearsonr(spearman_correlation_matrix[5,:], model_coeff[5,:])

print "correlation btw pearson and coeff"

print pearsonr(pearson_correlation_matrix[0,:], model_coeff[0,:])
print pearsonr(pearson_correlation_matrix[1,:], model_coeff[1,:])
print pearsonr(pearson_correlation_matrix[2,:], model_coeff[2,:])
print pearsonr(pearson_correlation_matrix[3,:], model_coeff[3,:])
print pearsonr(pearson_correlation_matrix[4,:], model_coeff[4,:])
print pearsonr(pearson_correlation_matrix[5,:], model_coeff[5,:])
```
